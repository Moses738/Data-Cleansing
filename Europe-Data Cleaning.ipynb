```python
Segment=[]
Sub_Types=[]
Values=[]
Date=[]
for i in df.columns[1:]:
    for j in df.index[3:]:
        Segment.append(df[i][0])
        Sub_Types.append(df[i][2])
        Values.append(df[i][j])
        Date.append(df[0][j])
```

```python
europe=pd.DataFrame({'Segment':Segment,'SubType':Sub_Types,'Values':Values,'Date':Date})
```

```python
europe
```

```python
r=re.compile('\w+')
r.search(europe['Segment'][0]).group()
```

```python
europe['Trail']=europe['Segment'].apply(lambda x:re.search('\w+',x).group())
```

```python
europe
```

```python
k=re.compile('\-\D+')
k.search(europe.Segment[0]).group().replace('-','')
```

```python
europe['Period']=europe['Segment'].apply(lambda x:re.search('\-\D+',x).group().replace('-',''))
```

```python
europe['Segment']=europe['Trail']
```
